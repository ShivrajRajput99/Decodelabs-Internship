## Objective

The objective of this project is to clean and prepare a raw dataset by identifying missing values, handling null values, removing duplicates, and correcting data formats. The cleaned dataset should be reliable and ready for further analysis.

---

## Tools Used

* Python
* Pandas Library
* Visual Studio Code (VS Code)

---

## Dataset Overview

The dataset contains 1200 records and 14 columns related to online sales transactions.

---

## Data Cleaning Steps

### 1. Missing Value Detection

The dataset was checked for missing values using the `isnull()` function.

#### Observation:

* CouponCode column contained 309 missing values.
* Other columns had no missing values.

#### Solution:

Missing values in the CouponCode column were replaced with **"No Coupon"**.

---

### 2. Duplicate Record Check

The dataset was examined for duplicate rows using the `duplicated()` function.

#### Observation:

* No duplicate records were found.

---

### 3. Duplicate Order ID Verification

The OrderID column was checked to ensure uniqueness.

#### Observation:

* No duplicate Order IDs were found.

---

### 4. Date Format Validation

The Date column was converted into datetime format using Pandas.

#### Observation:

* All dates were valid.
* No incorrectly formatted dates were detected.

---

### 5. Text Standardization

Text columns were standardized by:

* Removing extra spaces
* Converting text into proper title case

---

### 6. Data Type Verification

All columns were checked to ensure appropriate data types for analysis.
